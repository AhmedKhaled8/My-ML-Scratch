2. [What are the principles of visualization](https://www.sciencedirect.com/science/article/pii/S2666389920301896) ?

* Diagram First: before you make a visual, prioritize the information you want to share, envision it, and design it.

* Use an Effective Geometry and Show Data: Geometries are the shapes and features that are often synonymous with a type of figure; for example, the bar geometry creates a bar plot.

````python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
````

In matplolib we have:

- figure: which is the window that holds all plot elements.
- axes (subplots): which contain most figure elements. They are the drawing areas with coordinate system.

_Each figure has one axes by default but it can have many axes._


You can create a figure with its axes by many ways:

- Automatically: the library creates the figure when using any plot commands.


````python
x = np.arange(1, 10)
y = x * 2
plt.plot(x, y)
````

- `ax = plt.subplot(n, m , i)`: in a n \* m axes grid. Use the one at index (1-based index) to plot.


````python
ax11 = plt.subplot(2, 2, 1)
ax12 = plt.subplot(2, 2, 2)
ax13 = plt.subplot(2, 2, 3)
ax14 = plt.subplot(2, 2, 4)

ax11.plot(x, y)
````

- `fig, axes = plt.subplots(n, m)`: creates a figrue with an n rows \* m columns grid of axes. `axes[x, y]` to access the axes you want to plot on.


````python
fig, axes = plt.subplots(2, 2)
ax11, ax12, ax13, ax14 = axes.ravel()

ax12.plot(x, y)
````

In `matplotlib`, we can use either of 2 interfaces:

- stateful interface: applies to current figure and axes.


````python
sin = np.sin(np.linspace(-4, 4, 1000))
plt.subplot(2, 2, 1)
plt.plot(sin) # stateful interface
plt.subplot(2, 2, 2)
plt.plot(sin, c='r') # stateful interface
````

- object-oriented interface: explicitly use that object.


````python
fig, axes = plt.subplots(2, 2)
axes[1, 1].plot(sin, c='b') # object-oriented interface
````

Some different between the 2 interfaces:

|      stateful interface       |      Object-oriented interface      |
| :---------------------------: | :---------------------------------: |
|         `plt.title()`         |          `ax.set_title()`           |
|   `plt.xlim()` `plt.ylim()`   |   `ax.set_xlim()` `ax.set_ylim()`   |
| `plt.xlabel()` `plt.ylabel()` | `ax.set_xlabel()` `ax.set_ylabel()` |
| `plt.xticks()` `plt.yticks()` | `ax.set_xticks()` `ax.set_yticks()` |

`plot` is used for linear plotting. You can pass the x and y data and it will plot them and connect between them. Or, you can pass one data and it will plot it against sequential data from 0 to the count of the data passed - 1.

You can also pass some arguments to modify the plot regarding its style, its color, etc.


````python
fig, ax = plt.subplots(2, 4, figsize=(20, 5))
ax[0, 0].plot(sin)
ax[0, 1].plot(range(1000), sin)
ax[0, 2].plot(np.linspace(-4, 4, 1000), sin)
ax[0, 3].plot(sin[::100], 'o') # set the plot to disconnected dots
ax[1, 0].plot(sin, c = 'r') # set the color of the plot to red
ax[1, 1].plot(sin, '--') # set the plot to dashed lines
ax[1, 2].plot(sin, lw = 3) # set the line width of the plot to 3 (bold)
ax[1, 3].plot(sin[::100], '--o')
plt.tight_layout() # makes plot and legend fit well with each other - not overlapping
````

`scatter`

`scatter` plots continous quantities where each data is represented as a single not-connected point.

`plt.scatter(x, y) # scatter plot between x, y using default settings`

```
plt.scatter(x, y, z)
# scatter plot between x, y and map the size of the marker using the corresponding value of z
```

`plt.scatter(x, y, c = 'b') # scatter plot between x, y in blue markers`

`plt.scatter(x, y, marker = 'v') # scatter plot between x, y in triangle-down-like markers` [Scatter Markers](https://matplotlib.org/stable/api/_as_gen/matplotlib.markers.MarkerStyle.html#matplotlib.markers.MarkerStyle)


`plt.scatter(x, y, cmap = ""`

````python
x = np.random.uniform(size = 50)
y = x + np.random.normal(0, 0.1, size = 50)

fig, ax = plt.subplots(2, 2, figsize=(7, 7), subplot_kw = {'xticks': (), 'yticks': ()})

ax[0, 0].scatter(x, y)
ax[0, 0].set_title('scatter')

ax[0, 1].plot(x, y, 'o')
ax[0, 1].set_title('plot')

ax[1, 0].scatter(x, y, c = x - y, cmap = 'bwr', edgecolor = 'k')
ax[1, 1].scatter(x, y, c = x - y, cmap = 'bwr', edgecolor = 'k', s = np.abs(np.random.normal(scale = 20, size = 50)))

plt.tight_layout()
````

`bar`

`bar` allows us to plot a bar plot for categorial data.
`barh` allows to plot the bar plot in horizontal direction.

````python
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.bar(x, y)
plt.xlabel("x")
plt.ylabel("y")

plt.subplot(1, 2, 2)
plt.barh(x, y, color ='r')
plt.xlabel("x")
plt.ylabel("y")
````

**heatmaps**

We can use `imshow` to plot heatmaps.

````python
fig, ax = plt.subplots(2, 2, figsize = (15, 15))

arr = np.random.normal(size = (10, 10))

im1 = ax[0, 0].imshow(arr) # we will use im1 for the color bar
im2 = ax[0, 1].imshow(arr, interpolation = 'bilinear')
im3 = ax[1, 0].imshow(arr, cmap = 'gray')
im4 = ax[1, 1].imshow(arr, cmap = 'bwr', vmin = -1.5, vmax = 1.5)


plt.colorbar(im1, ax = ax[0, 0]) # build a color bar and assign it to the axes of ax[0, 0]
plt.colorbar(im2, ax = ax[0, 1])
plt.colorbar(im3, ax = ax[1, 0])
plt.colorbar(im4, ax = ax[1, 1])
````

`hexbin`

`hexbin` is useful for plotting overlapping data where the color of each point represent the number of data points overlapping at that position.

````python
x = np.random.normal(size=50000)
y = (x * 3 + np.random.normal(size=50000)) * 5
 

plt.figure(figsize=(10, 10))
plt.hexbin(x, y)
plt.colorbar()
plt.show()
````

`twinx`, `twiny`

If you tried to run the next code

````python
plt.figure(figsize=(10, 10))

x1 = np.arange(0, 11)
y1 = x1 + np.random.normal(size = 11)
y2 = x1 + np.random.normal(size = 11) + 100

plt.plot(x1, y1)
plt.plot(x1, y2)
````

Both `y1` and `y2` have very different ranges. If you want to study the correlation between quantities that vary in their ranges with fixed third quantity, it would be helpful to use `twinx` (or `twiny` but `twinx` is more common)

````python
x1 = np.arange(0, 11)
y1 = x1 + np.random.normal(size = 11)
y2 = x1 + np.random.normal(size = 11) + 10 ** 9

fig = plt.figure(figsize=(10, 10))

ax1 = fig.add_subplot(111) # get the current axes
ax1.plot(x1, y1, label="y1") # use label for legend
ax1.set_ylabel('y1')

ax2 = ax1.twinx()
ax2.plot(x1, y2, c = 'r', label="y2")
ax2.set_ylabel('y2')

fig.legend(loc="upper left", bbox_to_anchor=(0, 1), bbox_transform=ax1.transAxes) # to show the legend in twinx figure
````

3D Scattering

````python
from sklearn.datasets import load_iris

iris = load_iris()
x, y = iris.data, iris.target

fig = plt.figure(figsize=(10, 10))
ax = plt.axes(projection='3d')

# for loop to add label, legend
for i in range(3):
    ax.scatter(x[y==i, 1], x[y==i, 2], x[y==i, 3], label = i)

ax.legend(loc='upper left')
````
