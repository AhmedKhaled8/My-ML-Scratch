# Visualization


1. [Why visualization](https://analytiks.co/importance-of-data-visualization/) ?

* Correlations in Relationships: Without data visualization, it is challenging to identify the correlations between the relationship of independent variables. By making sense of those independent variables, we can make better business decisions. 

* Trends Over Time: While this seems like an obvious use of data visualization, it is also one of the most valuable applications. Itâ€™s impossible to make predictions without having the necessary information from the past and present. Trends over time tell us where we were and where we can potentially go. 

* Frequency: Closely related to trends over time is frequency. By examining the rate, or how often, customers purchase and when they buy gives us a better feel for how potential new customers might act and react to different marketing and customer acquisition strategies. 

2. [What are the principles of visualization](https://www.sciencedirect.com/science/article/pii/S2666389920301896) ?

* Diagram First: before you make a visual, prioritize the information you want to share, envision it, and design it.

* Use the Right Software: In other words, it might be unrealistic to expect complex, technical, and effective figures if you are using a simple spreadsheet program or some other software that is not designed to make complex, technical, and effective figures.

* Use an Effective Geometry and Show Data: Geometries are the shapes and features that are often synonymous with a type of figure; for example, the bar geometry creates a bar plot.


* always use **preceptual uniform sequential colormap** and avoid using **miscellaneous colormaps**.
* **Diverging** are very good if you are dealing with a scale that the end points are opposite and the middle is neutral.
* **Sequential** are great for scaling variable on a continous scale.

In matplolib we have:

_Each figure has one axes by default but it can have many axes._


You can create a figure with its axes by many ways:

- Automatically: the library creates the figure when using any plot commands.


````python
x = np.arange(1, 10)
y = x * 2
plt.plot(x, y)
````

- `ax = plt.subplot(n, m , i)`: in a n \* m axes grid. Use the one at index (1-based index) to plot.


````python
ax11 = plt.subplot(2, 2, 1)
ax12 = plt.subplot(2, 2, 2)
ax13 = plt.subplot(2, 2, 3)
ax14 = plt.subplot(2, 2, 4)

ax11.plot(x, y)
````

- `fig, axes = plt.subplots(n, m)`: creates a figrue with an n rows \* m columns grid of axes. `axes[x, y]` to access the axes you want to plot on.


````python
fig, axes = plt.subplots(2, 2)
ax11, ax12, ax13, ax14 = axes.ravel()

ax12.plot(x, y)
````

***Note:*** We can use the keyword argument `sharey = True` to make the y-axis limits the same for all subplots.

In `matplotlib`, we can use either of 2 interfaces:

- stateful interface: applies to current figure and axes.


````python
sin = np.sin(np.linspace(-4, 4, 1000))
plt.subplot(2, 2, 1)
plt.plot(sin) # stateful interface
plt.subplot(2, 2, 2)
plt.plot(sin, c='r') # stateful interface
````

- object-oriented interface: explicitly use that object.


````python
fig, axes = plt.subplots(2, 2)
axes[1, 1].plot(sin, c='b') # object-oriented interface
````

Some different between the 2 interfaces:

|      stateful interface       |      Object-oriented interface      |
| :---------------------------: | :---------------------------------: |
|         `plt.title()`         |          `ax.set_title()`           |
|   `plt.xlim()` `plt.ylim()`   |   `ax.set_xlim()` `ax.set_ylim()`   |
| `plt.xlabel()` `plt.ylabel()` | `ax.set_xlabel()` `ax.set_ylabel()` |
| `plt.xticks()` `plt.yticks()` | `ax.set_xticks()` `ax.set_yticks()` |

`plot` is used for linear plotting. You can pass the x and y data and it will plot them and connect between them. Or, you can pass one data and it will plot it against sequential data from 0 to the count of the data passed - 1.

You can also pass some arguments to modify the plot regarding its style, its color, etc.

* `linestyle`: the style of the lines connecting between the points.
* `color`: the color of the plot

````python
fig, ax = plt.subplots(2, 4, figsize=(20, 5))
ax[0, 0].plot(sin)
ax[0, 1].plot(range(1000), sin)
ax[0, 2].plot(np.linspace(-4, 4, 1000), sin)
ax[0, 3].plot(sin[::100], 'o') # set the plot to disconnected dots
ax[1, 0].plot(sin, c = 'r') # set the color of the plot to red
ax[1, 1].plot(sin, '--') # set the plot to dashed lines
ax[1, 2].plot(sin, lw = 3) # set the line width of the plot to 3 (bold)
ax[1, 3].plot(sin[::100], '--o')
plt.tight_layout() # makes plot and legend fit well with each other - not overlapping
````

`twinx`, `twiny`

If you tried to run the next code

````python
plt.figure(figsize=(10, 10))

x1 = np.arange(0, 11)
y1 = x1 + np.random.normal(size = 11)
y2 = x1 + np.random.normal(size = 11) + 100

plt.plot(x1, y1)
plt.plot(x1, y2)
````

Both `y1` and `y2` have very different ranges. If you want to study the correlation between quantities that vary in their ranges with fixed third quantity, it would be helpful to use `twinx` (or `twiny` but `twinx` is more common)

````python
x1 = np.arange(0, 11)
y1 = x1 + np.random.normal(size = 11)
y2 = x1 + np.random.normal(size = 11) + 10 ** 9

fig = plt.figure(figsize=(10, 10))

ax1 = fig.add_subplot(111) # get the current axes
ax1.plot(x1, y1, label="y1") # use label for legend
ax1.set_ylabel('y1')

ax2 = ax1.twinx()
ax2.plot(x1, y2, c = 'r', label="y2")
ax2.set_ylabel('y2')

fig.legend(loc="upper left", bbox_to_anchor=(0, 1), bbox_transform=ax1.transAxes) # to show the legend in twinx figure
````

`relplot`

`relplot` is a function of the `seaborn` library that allows us to plot relational plots bewteen quantative variables. These relational plots include scatter plot and line plot with different customizations to each type can be set inside `relplot`.

````python
student_data = pd.read_csv("assets/student-alcohol-consumption.csv")
student_data.head()
````

### Scatter Plot

```python
sns.relplot(
    data = data, # the pandas dataframe
    x = "xaxis", # a string represent the column of the data to be plotted on the x axis
    y = "yaxis", # a string represent the column of the data to be plotted on the y axis
    kind = "scatter", # a string represent the type of the relational plot 
)
```

````python
sns.relplot(
    data = student_data,
    x = "absences",
    y = "G3",
    kind = "scatter"
)
````

We can subclass the data based on other categorial column using `hue` keyword.

````python
sns.relplot(
    data = student_data,
    x = "absences",
    y = "G3",
    hue = "study_time",
    kind = "scatter"
)
````

We can also make separate the hue classes each in its own subplot instead of putting all of them together in a single plot. We can use the `col` or `row` keyword with `col_wrap` or `row_wrap` to define a max number of subplots per column or row.

````python
sns.relplot(
    data = student_data,
    x = "absences",
    y = "G3",
    col = "study_time",
    col_wrap = 2,
    col_order = ["<2 hours", "2 to 5 hours", "5 to 10 hours", ">10 hours"],
    kind = "scatter"
)
````

We can create a 2-factor subplots, factor by columns and factor by rows.

````python
sns.relplot(
    data = student_data,
    x = "G1",
    y = "G3",
    kind = "scatter",
    col = "schoolsup",
    col_order = ["yes", "no"],
    row = "famsup",
    row_order = ["yes", "no"]
)
````

We can use a continous (or ordinal) variable as the size of the markers by providing the column name in the `size` keyword.

````python
mpg = pd.read_csv("assets/mpg.csv")
mpg.head()
````

````python
sns.relplot(
    data = mpg,
    x = "horsepower",
    y = "mpg",
    kind = "scatter",
    size = "cylinders"    
)
````

We can also use `cylinders` gradient as the hue color of the markers.

````python
sns.relplot(
    data = mpg,
    x = "horsepower",
    y = "mpg",
    kind = "scatter",
    size = "cylinders",
    hue = "cylinders"  
)
````

For nominal categorial variable, it's preferred to use the marker shape and descriminative color to differntiate between the different classes.

````python
sns.relplot(
    data = mpg,
    x = "acceleration",
    y = "mpg",
    kind = "scatter",
    style = "origin",
    hue = "origin"  
)
````

### Line plot

We will still use `relplot` but we will change `kind` keyword argument to be `"line"` instead of `"scatter"`.

````python
sns.relplot(
    data = mpg,
    kind = "line",
    x = "model_year",
    y = "mpg"
)
````

The generated figure is a line plot. The solid blue line is the mean of the mpg for each model year. Where by default, seaborn generates a 95% confidence interval with the shaded blue, assuming the data is normally random sample.

We can customize the line plot as following:

````python
sns.relplot(
    data = mpg,
    kind = "line",
    x = "model_year",
    y = "horsepower",
    style = "origin",
    hue = "origin"
)
````

We can change the confidence interval interpretation to be the standard deviation of the values for each x-axis value through the `ci` keyword argument.

````python
sns.relplot(
    data = mpg,
    kind = "line",
    x = "model_year",
    y = "horsepower",
    style = "origin",
    hue = "origin",
    ci = "sd" # stands for standard deviation
)
````

Or remove it and keep the mean only.

````python
sns.relplot(
    data = mpg,
    kind = "line",
    x = "model_year",
    y = "horsepower",
    style = "origin",
    hue = "origin",
    ci = None # stands for standard deviation
)
````

We can show the markers of the points using `markers` keyword argument and keep the lines solid not dashed using `dashes` keyword argument.

````python
sns.relplot(
    data = mpg,
    kind = "line",
    x = "model_year",
    y = "horsepower",
    style = "origin",
    hue = "origin",
    ci = None,
    dashes = False,  # to make all lines solid
    markers = True,  # to show the markers 
)
````

`catplot`

`catplot` is used for plots including a categorical variable. It includes count plots and bar plots.

````python
survey_data = pd.read_csv("assets/young-people-survey-responses.csv")

survey_data.head()
````

### Count plots

Plots used to show the number of observations of classes of a categorial variable.

````python
sns.catplot(
    data = survey_data,
    x = "Internet usage",
    kind = "count"
)
````

We can make the bars appear horizontally instead of the default vertical view by just naming the categorial column in the `y` keyword argument instead of `x`.

````python
sns.catplot(
    data = survey_data,
    y = "Internet usage",
    kind = "count"
)
````

We can also separate the subplots based on another cateogiral variable using either `col` or `row` keyword arguments.

````python
sns.catplot(
    data = survey_data,
    y = "Internet usage",
    kind = "count",
    col = "Gender"
)
````

### Bar plot

When we use bar plot with a categorial variable vs a continuous variable, it shows the mean of the continuous variable for each class in the categorical variable.

````python
sns.catplot(
    data = survey_data,
    x = "Gender",
    y = "Mathematics",
    kind = "bar"
)
````

We can specify the order of the classes' bars in the bar (or count) plot using `order` keyword argument.

````python
sns.catplot(
    data = student_data,
    x = "study_time",
    y = "G3",
    kind = "bar",
    order = ["<2 hours", "2 to 5 hours", "5 to 10 hours", ">10 hours"]
)
````

The black lines in the top of the bar (like in `relplot`) represent (by default) the confidence intervals of the continuous variable for each class of the categorical variable. We can modify it (like `relplot`) using the `ci` keyword argument.

````python
sns.catplot(
    data = student_data,
    x = "study_time",
    y = "G3",
    kind = "bar",
    order = ["<2 hours", "2 to 5 hours", "5 to 10 hours", ">10 hours"],
    ci = None
)
````

### Box plot

A box plot shows the distribution of a quantitative data for each class of the categorial variable. The box itself shows the 25th and 75th percentiles. The line in the middle of the box represents the median. The whiskers represent the remaining of the distribution and the floating circles represent the outliers.

We create the box plot by specifiying the `kind` keyword argument in `catplot` to `"box"`.

````python
sns.catplot(
    data = student_data,
    x = "study_time",
    y = "G3",
    kind = "box",
    order = ["<2 hours", "2 to 5 hours", "5 to 10 hours", ">10 hours"]
)
````

We can also customize the box plot to:

* add subgroups of another categorial variable using color by `hue`.
* Remove the outliers (or modify them) using the `sym` argument. If we passed an empty string `""`, no outliers will be shown.
* Modify the location of the whiskers using `whis` keyword either:
  * passing a float number (e.g., `2.0`) and that will be the factor multiplied by the IQR (Q3 - Q1). Default is 1.5*IQR.
  * passing a list of 2 numbers (e.g., `[5, 95]`) and the whiskers will be at those percentiles.

````python
sns.catplot(
    data = student_data,
    x = "internet",
    y = "G3",
    kind = "box",
    hue = "location",
    sym = "",
    whis = [10, 95]
)
````

### Point plots

Point plot show the mean of a quantitative variable for each group of a categorical variable as a **point**. Vertical lines show 95% confidence intervals.

````python
sns.catplot(
    data = student_data,
    x = "famrel",
    y = "absences",
    kind = "point"
)
````

We can add "caps" to the end of the vertical lines using `capsize` keyword argument. If we set `capsize = 0.2`, it will draw caps (horizontal lines) to 20% of the distance between every 2 classes on the x-axis.

````python
sns.catplot(
    data = student_data,
    x = "famrel",
    y = "absences",
    kind = "point",
    capsize = 0.2
)
````

We can remove the lines connecting the points, we can do this by setting `join` to `False`.

````python
sns.catplot(
    data = student_data,
    x = "famrel",
    y = "absences",
    kind = "point",
    join = False
)
````

We can remove the vertical lines representing the confidence intervals (or change them) by setting `ci` to `None`.

````python
sns.catplot(
    data = student_data,
    x = "famrel",
    y = "absences",
    kind = "point",
    ci = None
)
````

We can as usual add another categorial variable as a hue.

````python
sns.catplot(
    data = student_data,
    x = "famrel",
    y = "absences",
    kind = "point",
    ci = None,
    hue = "school"
)
````

And we can change the estimator that points represent. By default, it's the mean of the quantitative variable for each class of the categorical variable, but, you can set a function that you want to use instead of mean. For example, the median from `np.median` by setting the `estimator` keyword argument.

````python
sns.catplot(
    data = student_data,
    x = "famrel",
    y = "absences",
    kind = "point",
    ci = None,
    hue = "school",
    estimator= np.median
)
````

### Add title

We can add a title to our seaborn plot by accessing the figure member of the plot object. The type of the plot is `<class 'seaborn.axisgrid.FacetGrid'>` which has a `fig` member that has the method `suptitle` that adds the title.

````python
g = sns.relplot(
    data = mpg,
    x = "weight",
    y = "horsepower",
    kind = "scatter"
)
print("Type of the plot is {}".format(type(g)))
g.fig.suptitle("Car Weight vs Horspower")
````

We can make the title higher by setting the `y` argument to a number higher than `1` which is the default. Setting it to `1.03` will make the title higer by 3%.

````python
g = sns.relplot(
    data = mpg,
    x = "weight",
    y = "horsepower",
    kind = "scatter"
)

g.fig.suptitle("Car Weight vs Horspower", y = 1.03)
````

### Setting x-label and y-label

We can set the x-label and y-label through the `set` method of the `FacetGrid` object by setting `xlabel` and `ylabel` to the new desired labels.

````python
g = sns.relplot(
    data = mpg,
    x = "weight",
    y = "horsepower",
    kind = "scatter"
)

g.fig.suptitle("Car Weight vs Horspower", y = 1.03)

g.set(xlabel = "Weight (kg)", ylabel = "Horspower (horse)")
````
